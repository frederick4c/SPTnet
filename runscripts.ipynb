```python
import numpy as np
import scipy.io as sio
import tifffile
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import argparse
from os.path import basename
from IPython.display import HTML

def visualize_inference(mat_path, tif_path, threshold=0.90):
    print(f"Loading results from {mat_path}...")
    results = sio.loadmat(mat_path)
    
    # Check shapes
    obj_est = results['obj_estimation'] # MATLAB GUI expects this to be [1, N, Q, T], my script saves [N, 1, Q, T] or [N, Q, T]
    xy_est = results['estimation_xy']   # [N, Q, T, 2]
    
    # Squeeze out batch and dummy dimensions so we get [Q, T] and [Q, T, 2]
    obj_est = np.squeeze(obj_est)
    xy_est = np.squeeze(xy_est)
    
    num_queries, num_frames = obj_est.shape
    
    print(f"Loading video from {tif_path}...")
    video = tifffile.imread(tif_path) # [T, H, W]
    
    # Network was trained on MATLAB .mat files read by h5py, which transposes (H, W) -> (W, H).
    # To match the network's native coordinate space, we MUST transpose the TIFF image,
    # just like MATLAB's Visualize_SPTnet_Outputs_GUI.m does with fliplr and rot90!
    
    fig, ax = plt.subplots(figsize=(6, 6))
    plt.subplots_adjust(left=0, bottom=0, right=1, top=1)
    ax.axis('off')
    
    img_plot = ax.imshow(np.zeros((video.shape[1], video.shape[2])), cmap='gray', vmin=0, vmax=255)
    scatter = ax.scatter([], [], c='r', s=20)
    
    scale = video.shape[2] / 2.0 # 32 for a 64x64 image
    
    def update(frame):
        # 1. Get raw frame and apply MATLAB-equivalent transformation (Transpose)
        raw_frame = video[frame]
        
        # MATLAB does: Iraw = fliplr(Iraw); Iraw = rot90(Iraw, 1);
        # This is exactly equivalent to a matrix transpose!
        disp_frame = np.transpose(raw_frame)
        
        # Update image data
        img_plot.set_data(disp_frame)
        img_plot.set_clim(vmin=np.min(disp_frame), vmax=np.max(disp_frame))
        
        # 2. Extract valid coordinates for this frame
        active_queries = obj_est[:, frame] > threshold
        
        if np.any(active_queries):
            # Coordinates are output as [-1, 1], so multiply by scale and add scale
            pts = xy_est[active_queries, frame, :]
            x = pts[:, 0] * scale + scale
            y = pts[:, 1] * scale + scale
            scatter.set_offsets(np.column_stack((x, y)))
        else:
            scatter.set_offsets(np.empty((0, 2)))
            
        return img_plot, scatter

    ani = animation.FuncAnimation(fig, update, frames=num_frames, interval=100, blit=True)
    return ani


ani = visualize_inference('Trained_models/inference_results/result_Example_testdata_000.mat', 'TestData/tiff_output/Example_testdata_000.tif', threshold=0.5)
HTML(ani.to_jshtml())
```
