```python
import tifffile
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# ==========================================
# 1. Setup & Load Data
# ==========================================
tif_path = 'TestData/tiff_output/Example_testdata_000.tif'
mat_path = 'Trained_models/inference_results/result_Example_testdata_000.mat'

video = tifffile.imread(tif_path)
results = sio.loadmat(mat_path)

obj_est = np.squeeze(results['obj_estimation'])
xy_est = np.squeeze(results['estimation_xy'])
H_est = np.squeeze(results['estimation_H'])
C_est = np.squeeze(results['estimation_C'])

num_frames = video.shape[0]
scale = video.shape[2] / 2
num_queries = obj_est.shape[0]
confidence_threshold = 0.90

# ==========================================
# 2. Setup Animation Canvas
# ==========================================
fig, ax = plt.subplots(figsize=(8, 8))
fig.subplots_adjust(left=0, bottom=0, right=1, top=1) # Remove white borders
im = ax.imshow(video[0], cmap='gray')
ax.axis('off')

# Placeholder for the red circles and text labels
scatter = ax.scatter([], [], s=100, edgecolors='red', facecolors='none', linewidths=2)
texts = []

# ==========================================
# 3. Animation Engine
# ==========================================
def animate(frame):
    im.set_data(video[frame]) # Update background
    
    # Clear old text labels
    for t in texts:
        t.remove()
    texts.clear()
    
    active_points = []
    
    # Find active particles
    for q in range(num_queries):
        if obj_est[q, frame] > confidence_threshold:
            x = xy_est[q, frame, 0] * scale + scale
            y = xy_est[q, frame, 1] * scale + scale
            active_points.append([x, y])
            
            # Add text
            txt = ax.text(x + 2, y - 2, f"H={H_est[q]:.2f}, D={C_est[q]*0.5:.2f}", 
                          color='red', fontsize=12, fontweight='bold')
            texts.append(txt)
            
    if active_points:
        scatter.set_offsets(np.array(active_points))
    else:
        scatter.set_offsets(np.empty((0, 2)))
        
    return [im, scatter] + texts

# Bake the animation (interval=200 means 5 frames per second)
anim = FuncAnimation(fig, animate, frames=num_frames, interval=200, blit=True)

# Prevent the static background frame from printing twice
plt.close(fig)

# ==========================================
# 4. Render as Inline Video!
# ==========================================
# This converts the animation into an interactive HTML/JS video player
HTML(anim.to_jshtml())
```
